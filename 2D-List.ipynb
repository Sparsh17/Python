## Row wise sum

```python
from sys import stdin

def rowWiseSum(mat, nRows, mCols):
    #Your code goes here
    li=[]
    for row in mat:
        sum=0
        for ele in row:
            sum = sum + ele
        li.append(sum)
    print(*li)
    
    # for i in range(nRows):
    #     sum=0
    #     for j in range(mCols):
    #         sum+=mat[i][j]
    #     print(sum,end=' ')
            



#Taking Input Using Fast I/O
def take2DInput() :
    li = stdin.readline().rstrip().split(" ")
    nRows = int(li[0])
    mCols = int(li[1])
    
    if nRows == 0 :
        return list(), 0, 0
    
    mat = [list(map(int, input().strip().split(" "))) for row in range(nRows)]
    return mat, nRows, mCols


#main
t = int(stdin.readline().rstrip())

while t > 0 :

    mat, nRows, mCols = take2DInput()
    rowWiseSum(mat, nRows, mCols)
    print()

    t -= 1
```

## Wave print

```python
from sys import stdin

def wavePrint(mat, nRows, mCols):
    #Your code goes here
    n=nRows
    m=mCols
    bFlag = 1
    j = 0
    
    while j < m:
        if bFlag == 1:
            for i in range(n):
                print(mat[i][j],end =' ')
            bFlag = 0
            j+=1
            
        else:
            for i in range(n-1, -1, -1):
                print(mat[i][j], end =' ')
            bFlag = 1
            j+=1




#Taking Input Using Fast I/O
def take2DInput() :
    li = stdin.readline().rstrip().split(" ")
    nRows = int(li[0])
    mCols = int(li[1])
    
    if nRows == 0 :
        return list(), 0, 0
    
    mat = [list(map(int, input().strip().split(" "))) for row in range(nRows)]
    return mat, nRows, mCols


#main
t = int(stdin.readline().rstrip())

while t > 0 :

    mat, nRows, mCols = take2DInput()
    wavePrint(mat, nRows, mCols)
    print()

    t -= 1
```

## Spiral print

```python
from sys import stdin

def spiralPrint(mat, nRows, mCols):
    #Your code goes here
    rows = nRows 
    cols = mCols 
    startRow, startCol, endRow, endCol = 0, 0, rows-1, cols-1
    while startRow<=endRow and startCol<=endCol:
        # Print startRow 
        for j in range(startCol, endCol+1): 
            print(mat[startRow][j], end=' ')
        startRow += 1
        if startRow>endRow or startCol>endCol: 
            break 
        # Print endCol 
        for i in range(startRow, endRow+1): 
            print(mat[i][endCol], end=' ') 
        endCol -= 1 
        if startRow>endRow or startCol>endCol: 
            break 
        for j in range(endCol, startCol-1, -1):
            print(mat[endRow][j], end=' ')
        endRow -= 1 
        if startRow>endRow or startCol>endCol:
            break 
        # Print startCol 
        for i in range(endRow, startRow-1, -1): 
            print(mat[i][startCol], end=' ') 
        startCol += 1 



#Taking Input Using Fast I/O
def take2DInput() :
    li = stdin.readline().rstrip().split(" ")
    nRows = int(li[0])
    mCols = int(li[1])
    
    if nRows == 0 :
        return list(), 0, 0
    
    mat = [list(map(int, input().strip().split(" "))) for row in range(nRows)]
    return mat, nRows, mCols


#main
t = int(stdin.readline().rstrip())

while t > 0 :

    mat, nRows, mCols = take2DInput()
    spiralPrint(mat, nRows, mCols)
    print()

    t -= 1
```
